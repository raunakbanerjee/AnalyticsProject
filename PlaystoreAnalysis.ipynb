# Data Manipulation

### I will be using median for missing values in NUMERICAL FORM
### Will be using mode for categorical data. male=1 and f=2 is a typr of cat data

```python
df.isnull().sum()
#So the rating column has 1474 values which are missing
```

```python
#Defining a function that will take it a series as I/P and fill all missing values with median value
def impute_median(series):
    return series.fillna(series.median())
```

```python
#So for the rating col I am replacing all the missing data using the median values
df.Rating=df['Rating'].transform(impute_median)
```

```python
df.isnull().sum()
# we can see that now there are 0 empty values in the rating column
```

```python
#Replacing categorical data using mode's of their respective columns

```

```python
#Finding mode values of all columns first
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
# Fill the missing categorical values with mode
df['Type'].fillna(str(df['Type'].mode().values[0]), inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]), inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]), inplace=True)
```

```python
#Now we can see there are no missing values
df.isnull().sum()
```

## Need to make all my other columns fully nummerical
1. Remove $ sign from price
2. Remove + and , sign from installs 

This is done using LAMBDA functions

```python
google_data=df
```

```python
### Let's convert Price, Reviews and Ratings into Numerical Values
google_data['Price'] = google_data['Price'].apply(lambda x: str(x).replace('$', '') if '$' in str(x) else str(x))
google_data['Price'] = google_data['Price'].apply(lambda x: float(x))
google_data['Reviews'] = pd.to_numeric(google_data['Reviews'], errors='coerce')
```

```python
google_data['Installs'] = google_data['Installs'].apply(lambda x: str(x).replace('+', '') if '+' in str(x) else str(x))
google_data['Installs'] = google_data['Installs'].apply(lambda x: str(x).replace(',', '') if ',' in str(x) else str(x))
google_data['Installs'] = google_data['Installs'].apply(lambda x: float(x))
```
